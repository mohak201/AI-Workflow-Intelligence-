```python
all_skills = (
    jobs_df["skills_required"]
    .dropna()
    .str.split(",")
    .explode()
    .str.strip()
)

top_skills = all_skills.value_counts()
top_skills.head(10)
```

```python
for skill in all_skills.unique():
    count = all_skills.str.lower().eq(skill.lower()).sum()
    print(f"{skill}: {count} occurrences")
```

```python
jobs_df["year_month"] = jobs_df["posting_date"].dt.to_period("M")
jobs_df["year_month"].head(5)
```

```python
for skill in top_skills.index:
    jobs_df[skill] = (
        jobs_df["skills_required"].astype(str).str.contains(skill, case=False, na=False, regex=False).astype(int)
    )
```

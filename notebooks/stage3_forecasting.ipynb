```python
jobs_df["posting_date"].dt.to_period("M").unique()
```

```python
all_skills = (
    jobs_df["skills_required"]
    .dropna()
    .str.split(",")
    .explode()
    .str.strip()
)

top_skills = all_skills.value_counts()
top_skills.head(10)
```

```python
for skill in all_skills.unique():
    count = all_skills.str.lower().eq(skill.lower()).sum()
    print(f"{skill}: {count} occurrences")
```

```python
jobs_df["year_month"] = jobs_df["posting_date"].dt.to_period("M")
jobs_df["year_month"].head(5)
```

```python
for skill in top_skills.index:
    jobs_df[skill] = (
        jobs_df["skills_required"].astype(str).str.contains(skill, case=False, na=False, regex=False).astype(int)
    )
```

```python
skill_freq = (
    jobs_df
    .groupby("year_month")[top_skills.index]
    .sum()
    .reset_index()
)

skill_freq
```

```python
python_ts = skill_freq[["year_month", "python"]].copy()
python_ts.columns = ["year_month", "demand"]
python_ts.head()
```

```python
skill_time_series = {}

for skill in top_skills.index:
    ts=skill_freq[["year_month", skill]].copy()
    ts.columns = ["year_month", "demand"]
    ts["year_month"] = ts["year_month"].dt.to_timestamp()

    skill_time_series[skill]=ts
```

```python
skill_time_series["python"].head()
```

```python
for skill, ts in skill_time_series.items():
    plt.figure(figsize=(12, 6))
    plt.plot(
        ts["year_month"],
        ts["demand"],
        marker="o",
        linewidth=2,
    )
    plt.title(f"Historical Demand for {skill}")
    plt.xlabel("Year")
    plt.ylabel("Job postings")
    plt.grid(True)
    plt.tight_layout()
    plt.savefig(
    f"../outputs/forecast/{skill}.png",
    dpi=300,
    bbox_inches="tight"
)
    plt.show()
```
